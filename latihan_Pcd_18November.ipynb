```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('dora.jpeg')

if img is None:
    print('Gagal membaca gambar. Pastikan path gambar benar.')
else:
    img2 = img.copy()  # Membuat salinan gambar untuk pengolahan selanjutnya
    cv2_imshow(img)

    rows, cols = img.shape[:2]

    # Proses ubah warna ke grayscale
    for i in range(rows):
        for j in range(cols):
            # Mendapatkan nilai warna pada setiap saluran (BGR)
            b = img.item(i, j, 0)
            g = img.item(i, j, 1)
            r = img.item(i, j, 2)

            # Menghitung rata-rata untuk menghasilkan citra grayscale
            gray = (b + g + r) // 3

            gray = min(255, max(0, gray))

            img[i,j] =gray
cv2_imshow(img)
```
